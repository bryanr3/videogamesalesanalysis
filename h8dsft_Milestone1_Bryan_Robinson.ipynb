Cleaning the Data

Dropping all the unused columns

```python
df_game = df_game.drop(['NA_Sales'], axis = 1)
df_game = df_game.drop(['EU_Sales'], axis = 1)
df_game = df_game.drop(['JP_Sales'], axis = 1)
df_game = df_game.drop(['Other_Sales'], axis = 1)
df_game = df_game.drop(['Critic_Count'], axis = 1)
df_game = df_game.drop(['User_Count'], axis = 1)
df_game = df_game.drop(['Developer'], axis = 1)
```

Dropping all Nan values from the columns

```python
df_game = df_game.dropna(subset = 'Critic_Score')
df_game = df_game.dropna(subset = 'User_Score')
df_game = df_game.dropna(subset = 'Rating')
df_game = df_game.dropna(subset = 'Year_of_Release')
df_game = df_game.dropna(subset ='Publisher')
```

Dropping all rows with  the 'tbd' value (which means that the game is not given a score yet)  in user score column

```python
df_game = df_game.drop(df_game[df_game['User_Score'] == 'tbd'].index)
```

Multiply the global sales column by 1 million 

```python
df_game['Global_Sales'] = df_game['Global_Sales'] * 1000000
```

Change the data type of user_score to float for descriptive statistics

```python
df_game['User_Score'] = df_game['User_Score'].astype(float)
df_game['Year_of_Release'] = df_game['Year_of_Release'].astype(int)
```

Tukeys rule to see the percentage of outliers for global sales column( global sales is not normally distributed)

```python
q1 = df_game['Global_Sales'].quantile(0.25)
q3 = df_game['Global_Sales'].quantile(0.75)
iqr = q3-q1

up = q3 + 1.5*iqr
down = q1 - 1.5*iqr

print(f'Upper Boundary: {up:.2f}' )
print(f'Lower Boundary: {down:.2f}')
```

```python
outlier = df_game['Global_Sales'][(df_game['Global_Sales']<down) | (df_game['Global_Sales']>up)]
```

```python
print(f'Percentage of outliers for Global Sales:{len(outlier)*100/len(df_game["Global_Sales"]):.2f}%')
```

There are a huge numbers of outliers in the global sales column, however this is normal in the video game industry. Most of the games that are sold very highly are from reputable publishers (Nintendo, Activision etc) and therefore they can have the advantage of a loyal fanbase and brand power.

Inferential Statistics

Are there any significant difference of the user scores of XBOX 1, PC and PS4 ?

H0: user scores of XBOX1, PS4 and PC video games are not significantly different

H1: user scores of XBOX1, PS4 and PC video games are significantly different

```python
df_game['Platform'].unique()
```

```python
df_game.info()
```

```python
score_xbox1 = df_game['User_Score'][(df_game['Platform'] == 'XOne')]
score_pc = df_game['User_Score'][(df_game['Platform'] == 'PC')]
score_ps4 = df_game['User_Score'][(df_game['Platform'] == 'PS4')]
```

```python
print(score_xbox1.mean())
print(score_xbox1.skew())
```

```python
sns.histplot(score_xbox1, kde = True, bins = 10)
```

 The Data for User_Score for Xbox 1 follows a near normal distribution

```python
print(score_pc.mean())
print(score_pc.skew())
```

```python
sns.histplot(score_pc, kde = True, bins = 10)
```

 The Data for User_Score for PC follows a near normal distribution

```python
print(score_ps4.mean())
print(score_pc.skew())
```

```python
sns.histplot(score_ps4, kde = True, bins = 10)
```

 The Data for User_Score for PS4 follows a near normal distribution

```python
res = stats.f_oneway(score_xbox1, score_pc, score_ps4)
print('P-value:',res.pvalue)
```

Since the p value is less than 0.05 (95% CI). Therefore, the null hypothesis is rejected and that the user score among xbox1, PC and PS4 Platform are significantly different
